```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constraints
def constraint1(x):
    return (12 - 2*x) / 3

def constraint2(x):
    return (8 - x) / 2

# Create an array of x values
x = np.linspace(0, 6, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$2x + 3y \leq 12$', color='blue')
plt.plot(x, constraint2(x), label=r'$x + 2y \leq 8$', color='green')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(constraint1(x), constraint2(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel('Product A (x)')
plt.ylabel('Product B (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

# Define the constraints
def constraint1(x):
    return (6 - x) / 2

def constraint2(x):
    return 5 - 2*x

# Create an array of x values
x = np.linspace(0, 3, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$x + 2y \leq 6$', color='blue')
plt.plot(x, constraint2(x), label=r'$2x + y \leq 5$', color='green')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(constraint1(x), constraint2(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.xlabel('Product X (x)')
plt.ylabel('Product Y (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

# Define the constraints
def constraint1(x):
    return (20 - 2*x) / 1

def constraint2(x):
    return (30 - 3*x) / 2

def constraint3(x):
    return (18 - x) / 2

# Create an array of x values
x = np.linspace(0, 10, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$2x + y \leq 20$', color='blue')
plt.plot(x, constraint2(x), label=r'$3x + 2y \leq 30$', color='green')
plt.plot(x, constraint3(x), label=r'$x + 2y \leq 18$', color='red')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(np.minimum(constraint1(x), constraint2(x)), constraint3(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 10)
plt.ylim(0, 15)
plt.xlabel('Product A (x)')
plt.ylabel('Product B (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

# Define the constraints
def constraint1(x):
    return (15 - x) / 2

def constraint2(x):
    return 20 - x

# Create an array of x values
x = np.linspace(0, 10, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$x + 2y \leq 15$', color='blue')
plt.plot(x, constraint2(x), label=r'$x + y \leq 20$', color='green')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(constraint1(x), constraint2(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 10)
plt.ylim(0, 15)
plt.xlabel('Product A (x)')
plt.ylabel('Product B (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# Define the constraints
def constraint1(x):
    return (12 - 3*x) / 4

def constraint2(x):
    return 6 - 2*x

# Create an array of x values
x = np.linspace(0, 4, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$3x + 4y \leq 12$', color='blue')
plt.plot(x, constraint2(x), label=r'$2x + y \leq 6$', color='green')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(constraint1(x), constraint2(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 4)
plt.ylim(0, 4)
plt.xlabel('Project P1 (x)')
plt.ylabel('Project P2 (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

# Define the constraints
def constraint1(x):
    return (8 - x) / 2

def constraint2(x):
    return (12 - 3*x) / 2

# Create an array of x values
x = np.linspace(0, 4, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$x + 2y \leq 8$', color='blue')
plt.plot(x, constraint2(x), label=r'$3x + 2y \leq 12$', color='green')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(constraint1(x), constraint2(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 4)
plt.ylim(0, 6)
plt.xlabel('Chocolate Cakes (x)')
plt.ylabel('Vanilla Cakes (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

# Define the constraints
def constraint1(x):
    return (18 - 3*x) / 4

def constraint2(x):
    return 10 - 2*x

# Create an array of x values
x = np.linspace(0, 6, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$3x + 4y \leq 18$', color='blue')
plt.plot(x, constraint2(x), label=r'$2x + y \leq 10$', color='green')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(constraint1(x), constraint2(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 5)
plt.ylim(0, 6)
plt.xlabel('Vehicle X (x)')
plt.ylabel('Vehicle Y (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

# Define the constraints
def constraint1(x):
    return (30 - 4*x) / 3

def constraint2(x):
    return (18 - x) / 2

def constraint3(x):
    return (24 - 3*x) / 2

# Create an array of x values
x = np.linspace(0, 6, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$4x + 3y \leq 30$', color='blue')
plt.plot(x, constraint2(x), label=r'$x + 2y \leq 18$', color='green')
plt.plot(x, constraint3(x), label=r'$3x + 2y \leq 24$', color='red')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(np.minimum(constraint1(x), constraint2(x)), constraint3(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 6)
plt.ylim(0, 8)
plt.xlabel('Product P1 (x)')
plt.ylabel('Product P2 (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region and Constraints")
plt.grid(True)
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# Define the constraints
def constraint1(x):
    return (5000 - 4000*x) / 3000

def constraint2(x):
    return (4500 - 2000*x) / 2500

def constraint3(x):
    return (3000 - 1000*x) / 1500

# Create an array of x values
x = np.linspace(0, 2, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$4000x + 3000y \leq 5000$ (Television)', color='blue')
plt.plot(x, constraint2(x), label=r'$2000x + 2500y \leq 4500$ (Print Media)', color='green')
plt.plot(x, constraint3(x), label=r'$1000x + 1500y \leq 3000$ (Social Media)', color='red')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(np.minimum(constraint1(x), constraint2(x)), constraint3(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 2)
plt.ylim(0, 3)
plt.xlabel('Campaign A Budget (x in thousands)')
plt.ylabel('Campaign B Budget (y in thousands)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region for Advertising Campaign Budget Allocation")
plt.grid(True)
plt.show()


# Define the constraints
def constraint1(x):
    return (30 - 2*x) / 4

def constraint2(x):
    return (24 - 3*x) / 2

def constraint3(x):
    return (20 - x) / 2

# Create an array of x values
x = np.linspace(0, 10, 100)

# Create plots for each constraint
plt.figure(figsize=(8, 6))
plt.plot(x, constraint1(x), label=r'$2x + 4y \leq 30$ (Meat)', color='blue')
plt.plot(x, constraint2(x), label=r'$3x + 2y \leq 24$ (Vegetables)', color='green')
plt.plot(x, constraint3(x), label=r'$x + 2y \leq 20$ (Rice)', color='red')

# Shade the feasible region
plt.fill_between(x, 0, np.minimum(np.minimum(constraint1(x), constraint2(x)), constraint3(x)), where=(x>=0), color='gray', alpha=0.5)

# Set up labels and limits
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel('Meal A (x)')
plt.ylabel('Meal B (y)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels for the constraints
plt.legend()

# Display the plot
plt.title("Feasible Region for Meal Planning in a School Cafeteria")
plt.grid(True)
plt.show()





```
